# 4. Renaming two-worded columns

```python
netflix=netflix.rename(columns={'Age Rating':'Age','IMDb Rating':'IMDb'}) #renames the Age & IMDb Rating  columns to Age and IMDb
netflix.info()
```

# 5. Lets answer some questions from this Netflix data

```python
total_mvs=netflix.Name.nunique() #returns the total number of movies in the dataset
print(f"The total movies are: {total_mvs}")
```

```python
total_yrs=netflix.Year.nunique() #returns the total years movies were released
print(f"The total release years are: {total_yrs}")
```

```python
total_cat=netflix.Category.nunique() #returns the total movie categories
print(f"Total movie categories: {total_cat}")
```

```python
total_age_rat=netflix.Age.nunique() #returns the number of unique age rating for the movies
print(f"Total Age ratings: {total_age_rat}")
```

```python
netflix['Duration']=netflix['Duration'].str.replace('mins ',' ')
netflix['Duration']=netflix['Duration'].astype(int)
netflix
```

```python
netflix.info()
```

```python
total_duration=netflix.Duration.sum() #returns the total minutes for all the movies
print(f"The total duration for all movies is: {total_duration}")
```

```python
year_movies=netflix.groupby('Year')['Name'].count().sort_values(ascending=True) #returns the number of movies released per year
year_movies
```

```python
year_movies.plot(kind='hist',
                 title='Total movies per release year',
                 xlabel='Years',
                 ylabel='Total movies',
                 bins=5,
                edgecolor='yellow'
                )
plt.xticks(rotation=45)
plt.show()
```

```python
movies_age=netflix.groupby('Age')['Name'].count() #returns total movies per age rating
movies_age
```

```python
movies_age.plot(kind='pie',
                title='Total Movies per age rating',
                autopct='%0.01f%%',
                startangle=90,
                explode=(0.15,0,0)
                )
plt.legend(title='Age rating',loc='upper right')
plt.show()
```

```python
movies_duration=netflix.groupby('Duration')['Name'].count().head(10) #Returns top 10 movies with the longest duration
movies_duration
```

```python
movies_duration.plot(kind='bar',
                     ylabel='Total Movies',
                     xlabel='Duration in Minutes',
                     title='Total movies by Duration',
                     color='yellow',
                     edgecolor='black'
                    )
plt.show()
```

```python
movies_cat=netflix.groupby('Category')['Name'].count().sort_values(ascending=False) #returns the total movies per category in a descending order
movies_cat
```

```python
movies_cat.plot(kind='line',
                title='Total movies per category',
                ylabel='Total Movies',
                xlabel='Category'
               )
plt.xticks(rotation=90)
plt.show()                
```

```python
movies_imdb=netflix.groupby('IMDb')['Name'].count() #returns the total movies per rating
movies_imdb
```

```python
movies_imdb.plot(kind='bar',
                title='Total movies per IMDb Rating',
                ylabel='Total Movies',
                xlabel='IMDb rating',
                 color='green',
                 ec='black'
               )
plt.xticks(rotation=90)
plt.show()    
```

```python
oldest_movie=netflix['Year'].min() #selects the Year column then returns the least year
print(f"The oldest movie was released in: {oldest_movie}")
```

```python
latest_movie=netflix['Year'].max() #selects the Year column the returns the largest year
print(f"The latest movie was released in: {latest_movie}")
```
